### 問1

自然数を入力とし、入力された自然数が3の倍数(Ex. 3, 6, 9, ...)であるときは`hoge`、3を含む数(Ex. 3, 13, 23, ...)であるときは`huga`、両方の条件を満たす時は`piyo`を出力する関数`nabeatsu`を定義してください

また、以下のテストケースを入力した際の出力結果を示してください

```
# テストケース1
6

# テストケース2
13

# テストケース3
33
```

````python
"""数値を受け取り、
- 3の倍数かつ'3'を含む → 'piyo'
- 3の倍数のみ → 'hoge'
- '3'を含むのみ → 'huga'
- どれでもない → '' を返す
"""
def nabeatsu(num: int) -> str:
    is_multiple_of_3 = (num % 3 == 0)
    contains_3 = ('3' in str(num))
    
    if is_multiple_of_3 and contains_3:
        return 'piyo'
    elif is_multiple_of_3:
        return 'hoge'
    elif contains_3:
        return 'huga'
    else:
        return ''

case_1 = 6
case_2 = 13
case_3 = 33

print('case_1:', nabeatsu(case_1))
print('case_2:', nabeatsu(case_2))
print('case_3:', nabeatsu(case_3))
````

### 問2

異なる自然数を含む配列とターゲットとなる自然数を入力として、配列に含まれる自然数を足し合わせてターゲットとなる自然数になる組み合わせを出力する関数`combination_sum`を定義してください

このとき、同じ自然数は何度でも利用してよいものとし、もし組み合わせが存在しない場合は空の配列を返してください

例えば、配列`[2, 3, 6, 7]`、ターゲット`7`を入力としたときの出力は`[[2,2,3],[7]]`になります

また、以下のテストケースを入力した際の出力結果を示してください

```
# テストケース1
配列: [2, 3, 6, 7]
ターゲット: 14

# テストケース2
配列: [2, 3, 5]
ターゲット: 8

# テストケース3
配列: [2, 9, 11]
ターゲット: 100
```

````python
def combination_sum(candidates: list[int], target: int) -> list[list[int]]:
    """
    candidates: 異なる自然数のリスト（重複なしを仮定）
    target: 目標となる自然数
    戻り値: 候補の和がtargetになる組み合わせのリスト（各組み合わせは非降順・重複なし）
    """
    # 結果を格納するリスト。見つかった組み合わせをここに追加する
    result: list[list[int]] = []
    
    # 候補を昇順にソートすることで、枝刈り（打ち切り）を効かせやすくする
    candidates = sorted(candidates)
    
    # 再帰探索用の内部関数を定義
    # start: 次に試す候補の開始インデックス（非降順を保つため）
    # current: 現在の組み合わせ（部分解）
    # current_sum: current に含まれる数の合計
    def backtrack(start: int, current: list[int], current_sum: int):
        # 合計がターゲットに到達したらコピーを結果に追加して戻る
        if current_sum == target:
            result.append(current[:])  # コピーを保存（参照ではなく値を保持）
            return
        
        # 合計がターゲットを超えたらこれ以上探索する必要はない（候補はソート済み）
        if current_sum > target:
            return
        
        # start から順に候補を試す
        for i in range(start, len(candidates)):
            val = candidates[i]
            
            # この値を足すとターゲットを超えるなら、それ以降の値も大きいのでループを打ち切る
            if current_sum + val > target:
                break
            
            # 候補を追加して再帰的に探索を続ける
            current.append(val)
            # 同じ数字を何度でも使えるため、次回の start は i のまま（i を再利用）
            backtrack(i, current, current_sum + val)
            # 探索後は元に戻す（バックトラック）
            current.pop()
    
    # 探索を開始（空の組み合わせ・合計0・start=0）
    backtrack(0, [], 0)
    # 見つかった組み合わせを返す
    return result
````

````python
candidates_1 = [2, 3, 6, 7]
target_1 = 14

candidates_2 = [2, 3, 5]
target_2 = 8

candidates_3 = [2, 9, 11]
target_3 = 100

print(combination_sum(candidates_1, target_1))
print(combination_sum(candidates_2, target_2))
print(combination_sum(candidates_3, target_3))
````

### 問3

括弧を含む文字列を入力として、括弧の内包関係が正しいかどうかを判定し、正しければ`True`、正しくなければ`False`を返す関数`format_checker`を定義してください

例えば、`[](){}`や`[({})]`は正しい、`[(]`や`{[(})]`は正しくないと判定します

また、以下のテストケースを入力した際の出力結果を示してください

```
# テストケース1
Hello I’m [(firstname) (lastname)] and I’m {age} years old!

# テストケース2
Hello I’m [(firstname) (lastname]) and I’m {age} years old!

# テストケース3
Hello I’m ((firstname) (lastname) and I’m {age} years old!
```

````python
def format_checker(s: str) -> bool:
    """
    括弧の内包関係が正しいかどうかを判定する関数
    
    Args:
        s: チェック対象の文字列
        
    Returns:
        bool: 括弧が正しく対応していればTrue、そうでなければFalse
    """
    
    # スタック（開き括弧を保存するリスト）
    stack = []
    
    # 括弧の対応関係を定義
    # キー 閉じ括弧、値: 対応する開き括弧
    bracket_map = {
        ')': '(',
        ']': '[',
        '}': '{'
    }
    
    # 文字列を1文字ずつ走査
    for char in s:
        # 開き括弧の場合
        if char in '([{':
            # スタックに追加（プッシュ）
            stack.append(char)
        
        # 閉じ括弧の場合
        elif char in ')]}':
            # スタックが空の場合、対応する開き括弧がない
            if not stack:
                return False
            
            # スタックの最後の要素を取り出す（ポップ）
            last_open = stack.pop()
            
            # 取り出した開き括弧と現在の閉じ括弧が対応しているかチェック
            if last_open != bracket_map[char]:
                return False
        
        # 括弧以外の文字は無視
    
    # すべての文字を処理した後
    # スタックが空なら、すべての括弧が正しく閉じられている
    # スタックに要素が残っていれば、閉じられていない括弧がある
    return len(stack) == 0
````

````python
case_1 = 'Hello I’m [(firstname) (lastname)] and I’m {age} years old!'
case_2 = 'Hello I’m [(firstname) (lastname]) and I’m {age} years old!'
case_3 = 'Hello I’m ((firstname) (lastname) and I’m {age} years old!'

print('case_1:', format_checker(case_1))
print('case_2:', format_checker(case_2))
print('case_3:', format_checker(case_3))
````
