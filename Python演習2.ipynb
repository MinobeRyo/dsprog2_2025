## 課題1

各問で示されている図を出力してください。

その際、以下の要件を必ず満たすようにしてください。


1.   for文を必ず用いること
2.   変数`size`の値を変えることで自由に図形の大きさを変更できること




### 問1

以下の図形を出力してください。

```
*
**
***
****
*****
```

````python
# 図形の大きさを設定
size = 5
def print_三角(size):
    for i in range(1, size + 1):
        print('*' * i)
print_三角(size)

# 空白行を挿入してみやすく
print()
size = 8
print_三角(size)
````

### 問2

以下の図形を出力してください。

```
*
 *
  *
   *
    *
```

````python
size = 5

def print_斜め(size):
    for i in range(size):
        # 各行でiの数だけスペースを入れて、その後に*を1つ出力する
        print(' ' * i + '*')

print_斜め(size)

print()
size = 8
print_斜め(size)
````

### 問3

以下の図形を出力してください。

```
    *
   *
  *
 *
*
```

````python
size = 5
def print_逆スラッシュ(size):
    for i in range(size):
        # 行が下に行くほどスペースが減り、*が右から左に移動するように
        spaces = size - i - 1
        print(' ' * spaces + '*')

print_逆スラッシュ(size)

print()
size = 8
print_逆スラッシュ(size)
````

### 問4

以下の図形を出力してください。

```
*   *
 * *
  *
 * *
*   *
```

````python
def print_X(size):
   
    for i in range(size):
        line = [' '] * size  # 空白で初期化
        #要は各行を空白文字で埋めた長さ5のリストを作成してそこに＊を配置していくイメージ
        
        # 左上から右下への対角線
        line[i] = '*'
        
        # 右上から左下への対角線
        line[size - 1 - i] = '*'
        
        print(''.join(line))

# 図形の大きさを設定 (奇数が望ましい)
size = 5
print_X(size)
print()
size = 8
print_X(size)
````

### 問5

以下の図形を出力してください。

```
    *
   ***
  *****
 *******
*********
```

````python
def print_ピラミッド(height):
    # 各行について処理
    for i in range(height):
        # 現在の行に必要な空白の数を計算
        spaces = height - i - 1
        
        # 現在の行に必要なアスタリスクの数を計算する(2*i + 1は奇数を生成)
        stars = 2 * i + 1
        
        # 空白とアスタリスクを連結して出力
        print(' ' * spaces + '*' * stars)

# ピラミッドの高さを設定
height = 5
print_ピラミッド(height)

print()
# ピラミッドの高さを設定
height = 8
print_ピラミッド(height)
````
